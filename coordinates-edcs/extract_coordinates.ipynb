```python
# CODE THAT WORKS TO MAKE XYZ FILES

# open files and create list of CID
f = open('DEDuCT_3D_structure.sdf').read()
df = pd.read_csv('/home/sguilber/Desktop/estrogen_similarity/SMARTS_classification/106EDCs_Smiles.csv')
#df = pd.read_csv('/home/sguilber/Desktop/estrogen_similarity/SMARTS_classification/106EDCs_Smiles.csv')
cid_df = df['Primary identifier'].str[4:]
cid_list = cid_df.tolist()
key = '$$$$'

# Text list that is a list of all coordinates
text = []

FILE_TYPES = {
    "sdf": {"num_atoms_line": 3, "split_func": lambda x: x.split()[:4]}
}

for cid in cid_list:
    first = f.find('$$$$\n' + cid + '\n')
    nxt = f.find('M  END', first+1)
    text1 = f[first:nxt].splitlines()[1:] # text1 is a list of coordinates for 1 molecule
    #text1 = text1[3:] # Removes the first 3 items of the list
    #text1.insert(0, '$$$$') # Adds separator $$$$ to separate molecules
    #text1.insert(1, cid) # Adds CID identifier after the separator
    
    with open('coordinates_sdf.sdf','w') as tfile:
        tfile.write('\n'.join(text1))
    
    with open('coordinates_sdf.sdf', 'r') as sdf:
        lines = sdf.readlines()
        num_atoms = int(lines[FILE_TYPES['sdf']["num_atoms_line"]][:3].strip())
        lines = lines[4:4+num_atoms]
        atoms = [FILE_TYPES['sdf']["split_func"](line) for line in lines if line.strip()]
        atoms = [line for line in atoms if line]
        for i in range(len(atoms)):
            atoms[i][0],atoms[i][1],atoms[i][2], atoms[i][3] = atoms[i][3],atoms[i][0],atoms[i][1],atoms[i][2]
        atoms2 = [' '.join(atom) for atom in atoms]
        atoms2.insert(0,'$$$$')
        atoms2.insert(1,str(num_atoms))
        atoms2.insert(2,cid)
        
    text = text + atoms2 # Combines all lists for all molecules

for l in text:
    print(l)

# Saves list as a txt file 
with open('f_coordinates.txt','w') as tfile:
    tfile.write('\n'.join(text))
```

```python
# Attempt 4
import re

START_PATTERN = '$$$$'
END_PATTERN = 'M  END'

with open('DEDuCT_3D_structure.sdf') as file:
    match = False
    newfile = None

    for line in file:
        if re.match(START_PATTERN, line):
            match = True
            newfile = open('my_new_file.txt', 'w')
            continue
        elif re.match(END_PATTERN, line):
            match = False
            newfile.close()
            continue
        elif match:
            newfile.write(line)
            newfile.write('\n')
```
