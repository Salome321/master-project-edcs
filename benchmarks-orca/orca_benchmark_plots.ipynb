```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def perfomance(N,fs):
    id = [N[i]*fs[0]/N[0] for i in range(len(N))]
    sp = [fs[0]**2/fs[i] for i in range(len(N))]
    eff = [100*sp[i]/id[i] for i in range(len(N))]
    return id,sp,eff
```

```python
#All orca comparing functinals. Single point energy calculations. Orca/SPE (Single Point Energy)
func = np.array(['PBE', 'TPSS', 'B3LYP', 'PBE0', 'TPSSH', 'B2PLYP'])
walltime_s_func = np.array([97, 150, 333, 331, 385, 446])
walltime_m_func = np.array([1.62, 2.5, 5.55, 5.52, 6.42, 7.43])
energy_func = np.array([ -923.90292, -925.21840, -924.511237, -923.99248, -925.123015, -924.1009769 ])
fig = plt.figure()
ax1 = plt.subplot(121)
ax1.grid()
plt.plot(func,energy_func,linestyle='-', color='b')
plt.ylabel("Energy (Hartrees)",fontsize=11, fontname="Arial")
plt.xlabel("Functional",fontsize=11, fontname="Arial")
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)

ax1 = plt.subplot(122)
ax1.grid()
plt.plot(func,walltime_m_func,linestyle='-', marker='o', color='red')
plt.ylabel("Walltime (minutes)",fontsize=11, fontname="Arial")
plt.xlabel("Functional",fontsize=11, fontname="Arial")
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)

fig.set_size_inches(15, 5.2)
plt.show()
```

```python
#Comparing basis sets for orca. Single Point Energy (SPE)
basis= np.array(['def2-SVP', 'def2-TZVP', 'aug-cc-pVDZ', 'aug-cc-pVTZ', 'cc_pVDZ', 'cc_pVTZ', 'ma_def2_SVP'])
walltime_s_basis = np.array([112, 298, 395, 327, 1031, 184, 419, 249, 9080, 39368])
walltime_m_basis = np.array([1.87, 4.97, 5.45, 17.18, 3.07, 6.98, 4.15])
energy_basis = np.array([ -923.20191, -924.194695, -923.992483, -924.183326, -923.92294, -924.167572, -923.24170])
fig = plt.figure()
ax1 = plt.subplot(121)
ax1.grid()
plt.plot(basis,energy_basis,linestyle='-', color='b')
plt.ylabel("Energy (Hartrees)",fontsize=11, fontname="Arial")
plt.xlabel("Basis Set",fontsize=11, fontname="Arial")
plt.xticks(fontsize = 11, rotation = 45)
plt.yticks(fontsize = 11)

ax1 = plt.subplot(122)
ax1.grid()
plt.plot(basis,walltime_m_basis,linestyle='-', marker='o', color='red')
plt.ylabel("Walltime (minutes)",fontsize=11, fontname="Arial")
plt.xlabel("Basis Set",fontsize=11, fontname="Arial")
plt.xticks(fontsize = 11, rotation = 45)
plt.yticks(fontsize = 11)

fig.set_size_inches(15, 5.2)
plt.show()
```

```python
#Comparing cores - Orca / Single Point Energy
cores= np.array([2,4,6,8,10,12,16,20])
walltime_s_cores = np.array([900, 554, 408, 331, 282, 238, 200, 175])
walltime_m_cores = np.array([17.25, 9.23, 6.8, 5.52, 4.7, 3.97, 3.33, 2.91])
energy_cores = np.array([ -923.992483, -923.99248, -923.99248,  -923.99248,  -923.99248, -923.99248, -923.992483, -923.992483])
id_cores,p_cores,eff_cores = perfomance(cores,walltime_m_cores)
fig = plt.figure()
ax1 = plt.subplot(121)
ax1.grid()
#plt.plot(cores,walltime_m_cores,linestyle='-', color='b')
plt.ylabel("Walltime (minutes)",fontsize=11, fontname="Arial")
plt.xlabel("Cores",fontsize=11, fontname="Arial")
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
plt.plot(cores,id_cores/id_cores[0],linestyle='-', color='b', label="Ideal Speed-Up")
plt.plot(cores,p_cores/p_cores[0],linestyle='--', marker='o', color='red', label='Actual Speed-Up')
plt.ylabel("Speed-up",fontsize=11, fontname="Arial")
plt.xlabel("cores",fontsize=11, fontname="Arial")
plt.legend()

ax1 = plt.subplot(122)
ax1.grid()
#plt.plot(cores,walltime_m_cores,linestyle='-', marker='o', color='red')
plt.axhline(y=50,c='b',linestyle='-', linewidth=1.5, label="Optimal Efficiency")
#plt.axvline(x=28,c='b',linestyle='-', linewidth=1.5)
plt.plot(cores,eff_cores,linestyle='--', marker='o', color='red', label='Actual Efficiency')
plt.ylabel("Efficiency %",fontsize=11, fontname="Arial")
plt.xlabel("Cores",fontsize=11, fontname="Arial")
plt.legend()
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)

fig.set_size_inches(15, 5.2)
plt.show()
```
